```python
from google_play_scraper import app #disini koding buat import
import pandas as pd
import numpy as np
```

```python
#scrape jumlah ulasan yang diinginkan
from google_play_scraper import Sort, reviews

result, continuation_token=reviews(
    'com.miHoYo.GenshinImpact', # untuk memenggil Genshin Impact dari google play
    lang='id', # untuk men scrape data ulasan aplikasi Genshin Impact yang berada di google play store
    country='id', # untuk setting bahasa nya menjadi bahasa indonesia
    sort=Sort.MOST_RELEVANT,# most_relevan untuk mendapatkan ulasan yang paling relevant
    count=1000, # jumlah ulasan yang mau di ambil
    filter_score_with=None#  None untuk mengambil semua score atau ratting bintang 1 sampai 5
)
```

```python
#Dataframe dengan nama
data = pd.DataFrame (np.array(result),columns=['review'])
data = data.join(pd.DataFrame(data.pop('review').tolist()))
data.head() #untuk me review
```

```python
len(data) #untuk mengecek data
```

```python
data = data[['userName','content','score','at']] # untuk memilih data yang mau diambil
data.head()
```

```python
data = data.rename(columns={'userName':'Nama','content':'Review','score':'Reting','at':'Data'}) # merubah nama data diatas
data.head()
```

```python
data.to_csv('Ulasan Genshin Impact.csv',index=False) # untuk mendownload data
```

```python
dataGGenshinImpact=pd.read_csv('Ulasan Genshin Impact.csv') #
dataGGenshinImpact.head()
```

# **`Pelabelan `**

```python
def pelabelan(score):
  if score >3:
    return 'Positive' # jika angka lebih dari tiga maka positiv
  if score ==3:
    return 'Netral' # jika angka tiga maka netral
  if score <3:
    return 'Negative' # jika angka kurang dari tiga maka negatif
dataGGenshinImpact['Label']=dataGGenshinImpact['Reting'].apply(pelabelan)
dataGGenshinImpact.head()
```

```python
dataGGenshinImpact.to_csv('Pelabelan Genshin Impact.csv',index=False) # untuk download
```
